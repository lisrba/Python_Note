### 試一下從dataframe抓資料進模型產出向量，輸出成CSV檔

```python
for i in range(300):
    print(PeiW2V[pos_words_df["word"][0]][i])
    print('\n')
    print(',')
```

```python
with open("C:"+os.sep+"temp"+os.sep+"pos_vec.csv", "a",encoding='utf-8') as filetest:
    for i in range(len(pos_words_df)):
        filetest.write(str(pos_words_df["word"][i]))
        filetest.write(',')
        for j in range(300):
            filetest.write(str(PeiW2V[pos_words_df["word"][i]][j]))
            filetest.write(',')
        filetest.write('\n')  
```

```python
with open("C:"+os.sep+"temp"+os.sep+"neg_vec.csv", "a",encoding='utf-8') as filetest:
    for i in range(len(neg_words_df)):
        filetest.write(str(neg_words_df["word"][i]))
        filetest.write(',')
        for j in range(300):
            filetest.write(str(PeiW2V[neg_words_df["word"][i]][j]))
            filetest.write(',')
        filetest.write('\n')  
```
